```python
df= pd.read_csv("CountryTable.csv", index_col=2)
df
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas
#Variables filtradas: Se filtrar√≥n los continentes(columna 2) "North America"
# y"South America" con su "population"
filtro4=df.loc[["North America","South America"], ["population"]]
filtro4
```

```python
#Filtro por cabecera
#Variables filtradas: Imprimimos las primeras 15 filas
filtro5=df.head(15)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
#Variables filtradas: Filtramos los datos que cumplan que en la
# columna "life_expectancy" tengan un valor mayor a 76
filtro6=df[df["life_expectancy"] > 76]
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
#Variables filtradas: Filtramos los datos que cumplan que en la
# columna "life_expectancy" tengan un valor mayor a 76 y que en la
# columna "region" tengan el dato "Caribbean"
filtro7=df[(df["life_expectancy"] > 76) & (df["region"] == "Caribbean")]
filtro7
```

```python
#Filtro o
#Variables filtradas: Filtramos los datos que cumplan que en la
# columna "life_expectancy" tengan un valor mayor a 76 o que en la
# columna "region" tengan el dato "Caribbean"
filtro8=df[(df["life_expectancy"] > 76) | (df["region"] == "Caribbean")]
filtro8
```

```python
#Filtro not
#Variables filtradas: Filtramos los datos que cumplan que en la
# columna "life_expectancy" no tengan un valor mayor a 76 y que en la
# columna "region" no tengan el dato "Caribbean"
filtro9=df[~(df["life_expectancy"] >= 76) & ~(df["region"] == "Caribbean")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Filtro9.csv")
```
